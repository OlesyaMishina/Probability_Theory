```python
plt.scatter(X1, Y1)
plt.plot(X1, b0+X1*b1)
plt.show
```

```python

X1 = np.array([30,30,40, 40, 20, 20, 50, 50, 10, 10, 60, 60]) 
Y1 = np.array([37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84]) 
b1=(np.mean(X1*Y1)-np.mean(X1)*np.mean(Y1))/(np.mean(X1**2)-np.mean(X1)**2)
b0=np.mean(Y1)-b1*np.mean(X1)
R=np.corrcoef(X1, Y1)[0, 1]**2
print(f'Коэффициент детенрминации {R}')
print(b0, b1)
plt.scatter(X1, Y1)
plt.plot(X1, b0+b1*X1)
plt.show
```

градиентный спуск:

```python
def mse(b0, b1, x, y, n):
        return np.sum((b0+b1*x)-y)**2/n
alpha=0.00001
b0=0.1
b1=0.1
n=len(X1)
for i in range (1000000*2):
    b0-=alpha*(2/n)*np.sum((b0+b1*X1)-Y1)
    b1-=alpha*(2/n)*np.sum(((b0+b1*X1)-Y1)*X1)
    if i%100000==0:
        print(f'i ={i}, mse = {mse}, b0 = {b0}, b1 = {b1}')
```

Функция в питон для линейной регрессии:

```python
from sklearn.linear_model import LinearRegression
X1 = np.array([30,30,40, 40, 20, 20, 50, 50, 10, 10, 60, 60]) 
Y1 = np.array([37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84]) 
x=X1.reshape(-1, 1)
x
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
model = LinearRegression()
model = LinearRegression()
model.fit(x, Y1)
r_sq = model.score(x, Y1)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)

```
