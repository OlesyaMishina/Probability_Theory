```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
a = np.array([12, 10, 11, 19, 13, 11, 17, 15, 19, 14, 21, 18, 21, 11, 17, 14, 15, 17, 20, 19]) 
b = np.array([11, 13, 18, 15, 17, 18, 10, 21, 26, 15, 11, 12, 15, 17, 10, 18, 18, 12,21, 20])
r=np.corrcoef(a,b)
print(r)
```

```python
alpha=0.05
pvalue = stats.ttest_ind(a,b)[1]
print(pvalue)
```

 Имея p-value из функции рассчитать наблюдаемое значение критерия.

```python
t_table = stats.t.ppf(1-alpha/2, len(a)+len(b)-2)
print(t_table)
pvalue_observed = stats.t.ppf(1-pvalue/2, len(a)+len(b)-2)
pvalue_observed
```
